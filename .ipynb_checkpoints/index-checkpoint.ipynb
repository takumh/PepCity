# Introduction

## Background and Motivation

Manchester City Football club is considered one of the best football teams in the world. They have won the English premier league four times in the last five seasons and are on track to win their fifth title. They have also won multiple domestic cups in England and the UK. Manchester city havent always been a successfull club however, and some fans argue that their recent success has only come as a result of ownership change, massive spending and their latest manager, Pep Guardiola. My goal was to investigate Manchester City's history in the EPL and compare their recent performances under Pep to previous managers.

```python
import pandas as pd
import re
import string
import matplotlib.pyplot as plt
import os, wordcloud
from datetime import datetime
from sklearn.linear_model import LinearRegression
import numpy as np
import random
```

```python
#hide
#standardise column names
def standardise_column_names(df, remove_punct=True):
    translator = str.maketrans(string.punctuation, ' '*len(string.punctuation))

    for c in df.columns:
        c_mod = c.lower()
        if remove_punct:            
            c_mod = c_mod.translate(translator)
        c_mod = '_'.join(c_mod.split(' '))
        if c_mod[-1] == '_':
            c_mod = c_mod[:-1]
        c_mod = re.sub(r'\_+', '_', c_mod)
        df.rename({c: c_mod}, inplace=True, axis=1)
    return df
```

# About the data

For my analysis, I used two datasets sourced from Kaggle in csv format. One contains all Premier League matches played since the EPL's inception in 1992 up to the recently ended 2022 season. The other dataset contains tweets related to all premier league teams sourced from 2020 for the months of July to October that capture most of the 2020 - 2021 season.

```python
#read in csv
matches = pd.read_csv('eplmatches.csv')

#standardise column names
standardise_column_names(matches)

#isolate records for manchester city - home and away
home_mask = matches["home"].str.contains("Manchester City")
away_mask = matches["away"].str.contains('Manchester City')

#create dfs for city - home and away and seperate for pep and before pep
Hcity_df = matches[home_mask]
before_pep_home = Hcity_df[Hcity_df['season_end_year'] < 2017]
pep_home = Hcity_df[Hcity_df['season_end_year'] >= 2017]

Acity_df = matches[away_mask]
before_pep_away = Acity_df[Acity_df['season_end_year'] < 2017]
pep_away = Acity_df[Acity_df['season_end_year'] >= 2017]
```

# Methodology

## Wrangling

The first task was to extract data relating to Manchester city from both the EPL dataset as well as the Tweets dataset. I then separated the data for home and away games and further separated it for games played before Pep Guardiola became coach (1992-2016) and for those played after he became coach (2016-2022). This was achieved using masks.

```python
#function to calculate percentages
def percentage(part, whole):
  return 100 * float(part)/float(whole)
```

```python
#count number of wins, draws and losses for all home games
Hdraws = sum(Hcity_df['ftr'].str.count('D'))
Hwins = sum(Hcity_df['ftr'].str.count('H'))
Hlosses = sum(Hcity_df['ftr'].str.count('A'))
Htotal = Hdraws + Hwins + Hlosses

Hperc_wins = percentage(Hwins,Htotal)
Hperc_draws = percentage(Hdraws,Htotal)
Hperc_losses = percentage(Hlosses, Htotal)

#count number of wins, draws and losses for all away games
Adraws = sum(Acity_df['ftr'].str.count('D'))
Awins = sum(Acity_df['ftr'].str.count('A'))
Alosses = sum(Acity_df['ftr'].str.count('H'))
Atotal = Adraws + Alosses + Awins

Aperc_wins = percentage(Awins,Atotal)
Aperc_draws = percentage(Adraws,Atotal)
Aperc_losses = percentage(Alosses, Atotal)

#tally totals for draws, wins and losses for all managers
draws = Hdraws + Adraws
wins = Hwins + Awins
losses = Hlosses + Alosses
total = Htotal + Atotal

perc_wins = percentage(wins,total)
perc_draws = percentage(draws,total)
perc_losses = percentage(losses, total)


#count number of wins, draws and losses for all home games before Pep
BPh_draws = sum(before_pep_home['ftr'].str.count('D'))
BPh_wins = sum(before_pep_home['ftr'].str.count('H'))
BPh_losses = sum(before_pep_home['ftr'].str.count('A'))
BPh_total = BPh_draws + BPh_wins + BPh_losses

BPh_perc_wins = percentage(BPh_wins, BPh_total)
BPh_perc_draws = percentage(BPh_draws, BPh_total)
BPh_perc_losses = percentage(BPh_losses, BPh_total)

#count number of wins, draws and losses for all away games before Pep
BPa_draws = sum(before_pep_away['ftr'].str.count('D'))
BPa_wins = sum(before_pep_away['ftr'].str.count('A'))
BPa_losses = sum(before_pep_away['ftr'].str.count('H'))
BPa_total = BPa_draws + BPa_losses + BPa_wins

BPa_perc_wins = percentage(BPa_wins, BPa_total)
BPa_perc_draws = percentage(BPa_draws, BPa_total)
BPa_perc_losses = percentage(BPa_losses, BPa_total)

#tally totals for draws, wins and losses for all managers
BP_draws = BPh_draws + BPa_draws
BP_wins = BPh_wins + BPa_wins
BP_losses = BPh_losses + BPa_losses
BP_total = BPa_total + BPh_total

BP_perc_wins = percentage(BP_wins, BP_total)
BP_perc_draws = percentage(BP_draws, BP_total)
BP_perc_losses = percentage(BP_losses, BP_total)

#count number of wins, draws and losses for all home games before Pep
Ph_draws = sum(pep_home['ftr'].str.count('D'))
Ph_wins = sum(pep_home['ftr'].str.count('H'))
Ph_losses = sum(pep_home['ftr'].str.count('A'))
Ph_total = Ph_draws + Ph_wins + Ph_losses

Ph_perc_wins = percentage(Ph_wins, Ph_total)
Ph_perc_draws = percentage(Ph_draws, Ph_total)
Ph_perc_losses = percentage(Ph_losses, Ph_total)

#count number of wins, draws and losses for all away games before Pep
Pa_draws = sum(pep_away['ftr'].str.count('D'))
Pa_wins = sum(pep_away['ftr'].str.count('A'))
Pa_losses = sum(pep_away['ftr'].str.count('H'))
Pa_total = Pa_draws + Pa_losses + Pa_wins

Pa_perc_wins = percentage(Pa_wins, Pa_total)
Pa_perc_draws = percentage(Pa_draws, Pa_total)
Pa_perc_losses = percentage(Pa_losses, Pa_total)

#tally totals for draws, wins and losses for peo
P_draws = Ph_draws + Pa_draws
P_wins = Ph_wins + Pa_wins
P_losses = Ph_losses + Pa_losses
P_total = Pa_total + Ph_total

P_perc_wins = percentage(P_wins, P_total)
P_perc_draws = percentage(P_draws, P_total)
P_perc_losses = percentage(P_losses, P_total)
```

Firstly, I wanted to look at the distribution of wins, losses and draws for Manchester City across all games in its Premier League history. The following plots show those results aggregated as percentages by home games, away games and the totals.

# xG via Machine Learning

To best learn how much Manchester City's goal margins have improved under Pep Guardiola, I trained a simple machine learning model to predict the goals expected to be scored and conceded by Manchester City given a specific opponent. I further specified that the model should predict differently for home and away games as we have seen that there are significant differences in performance for games played at home and those played away. To get an idea of the model's output, I selected 5 teams (Liverpool, Manchester United, Arsenal, Chelsea, Tottenham Hotspur),that have arguably been top teams historically in the Premier League and that would have faced Manchester City occasionally both before and after Pep Guardiola became manager. I also selected Watford, Southampton, Burnley, Aston Villa and Everton to add balance. The results are displayed below.

```python
#home games all
# Create df with aggregated goals scored and conceded by each opponent
opponent_goals = Hcity_df.groupby('away').agg({'homegoals': 'sum', 'awaygoals': 'sum', 'away': 'count'})
opponent_goals.rename(columns={'away': 'games_played'}, inplace=True)

#calculate average goals scored and conceded per game by each opponent
opponent_goals['avg_goals_scored'] = opponent_goals['homegoals'] / opponent_goals['games_played']
opponent_goals['avg_goals_conceded'] = opponent_goals['awaygoals'] / opponent_goals['games_played']

#Train linear regression model to predict expected goals scored and conceded by Manchester City against a given opponent
X = opponent_goals[['avg_goals_scored', 'avg_goals_conceded']]
y_scored = opponent_goals['avg_goals_scored']
y_conceded = opponent_goals['avg_goals_conceded']
model_scored = LinearRegression().fit(X, y_scored)
model_conceded = LinearRegression().fit(X, y_conceded)

#Use the trained model to make predictions for a given opponent
opponent = 'Arsenal'  # replace with the desired opponent
opponent_stats = opponent_goals.loc[opponent, ['avg_goals_scored', 'avg_goals_conceded']].values.reshape(1, -1)
expected_scored = model_scored.predict(opponent_stats)[0]
expected_conceded = model_conceded.predict(opponent_stats)[0]

#print(f'Expected goals scored by Man City against {opponent}: {expected_scored:.2f}')
#print(f'Expected goals conceded by Man City against {opponent}: {expected_conceded:.2f}')
```

```python
#Away games all
#Create a df with aggregated goals scored and conceded by each opponent
opponent_goals = Acity_df.groupby('home').agg({'homegoals': 'sum', 'awaygoals': 'sum', 'home': 'count'})
opponent_goals.rename(columns={'home': 'games_played'}, inplace=True)

#Calculate the average goals scored and conceded per game by each opponent
opponent_goals['avg_goals_scored'] = opponent_goals['awaygoals'] / opponent_goals['games_played']
opponent_goals['avg_goals_conceded'] = opponent_goals['homegoals'] / opponent_goals['games_played']

# Train a model to predict the expected goals scored and conceded by Manchester City against a given opponent
X = opponent_goals[['avg_goals_scored', 'avg_goals_conceded']]
y_scored = opponent_goals['avg_goals_scored']
y_conceded = opponent_goals['avg_goals_conceded']
model_scored = LinearRegression().fit(X, y_scored)
model_conceded = LinearRegression().fit(X, y_conceded)

# Use the trained model to make predictions for a given opponent
opponent = 'Arsenal'  # replace with the desired opponent
opponent_stats = opponent_goals.loc[opponent, ['avg_goals_scored', 'avg_goals_conceded']].values.reshape(1, -1)
expected_scored = model_scored.predict(opponent_stats)[0]
expected_conceded = model_conceded.predict(opponent_stats)[0]

#print(f'Expected goals scored by Man City against {opponent}: {expected_scored:.2f}')
#print(f'Expected goals conceded by Man City against {opponent}: {expected_conceded:.2f}')
```

## Tweet city: does Manchester City have no fans?

<img src="tweet.webp" alt="Manchester City Logo" width="400" height="300" align="center"/>

There is a general belief in the Premier League that Manchester City have no fans or have 'plastic fans'. On the other hand, Manchester City fans like myself believe that we are overly hated and spoken negatively about by rival fans. I decided to investigate these claims using nearly 140 000 tweets that were filtered for the hashtags : #ManchesterCity, #Mancity or #MCFC. The tweets in the dataset are from July to October in 2020 which captures the end of the pre-season and the first half of the season ending in 2021. I was also interested inseeing what the discourse about Manchester City in this season was and what the general sentiment carried by the tweets was. 

```python
#hide
#read in the epl tweets file
tweets1 = pd.read_csv('archive/df1.csv')
#tweets1.head()
tweets2 = pd.read_csv('archive/df2.csv')
#tweets2.head()

#isolate tweets for manchester city
tweet1_mask = tweets1["file_name"].str.contains("ManchesterCity")
tweet2_mask = tweets2["file_name"].str.contains("ManchesterCity")

#create data frames with tweets
city1 = tweets1[tweet1_mask]
city2 = tweets2[tweet2_mask]

#combine tweets into one df
tweet_frames = [city1, city2]
city_tweets = pd.concat(tweet_frames)
```

```python
#define function to assign labels
def label_polarity(score):
    if score > 0:
        return 'positive'
    elif score < 0:
        return 'negative'
    else:
        return 'neutral'

#apply function to create new column
city_tweets['polarity_label'] = city_tweets['polarity'].apply(lambda x: label_polarity(x))

#positive tweets
pos_tweets = city_tweets[city_tweets["polarity_label"] == 'positive']
#negative tweets
neg_tweets = city_tweets[city_tweets["polarity_label"] == 'negative']
#neutral tweets
neu_tweets = city_tweets[city_tweets["polarity_label"] == 'neutral']
```

```python
#create a date column by extracting date from `created_at`
city_tweets['date'] = pd.to_datetime(city_tweets['created_at']).dt.date

day_tweets = city_tweets.groupby('date').size().reset_index(name='counts')
pos_day_tweets = pos_tweets.groupby('date').size().reset_index(name='counts')
neg_day_tweets = neg_tweets.groupby('date').size().reset_index(name='counts')
neu_day_tweets = neu_tweets.groupby('date').size().reset_index(name='counts')
```

To truly capture the nature of the discourse about Manchester City, I also wanted to capture how the tweets frequency varied for the time period capture aggregated by sentiment. The plot below illustrates  this.

```python
import plotly.graph_objs as go
import plotly.offline as pyo
import plotly.io as pio
pio.renderers.default='notebook'

# create traces for each dataframe
trace1 = go.Scatter(x=day_tweets['date'], y=day_tweets['counts'], mode='lines', name='Total', line=dict(color='#1C2C5B'))
trace2 = go.Scatter(x=pos_day_tweets['date'], y=pos_day_tweets['counts'], mode='lines', name='Positive', line=dict(color = "#6CABDD"))
trace3 = go.Scatter(x=neg_day_tweets['date'], y=neg_day_tweets['counts'], mode='lines', name='Negative', line=dict(color = "#EC3325"))
trace4 = go.Scatter(x=neu_day_tweets['date'], y=neu_day_tweets['counts'], mode='lines', name='Neutral', line=dict(color='#D4A12A'))

# create a list of traces
data = [trace1, trace2, trace3, trace4]

# create the figure with the traces
fig = go.Figure(data=data)

fig.update_layout(
    title='Tweet Frequency Distribution',
    xaxis_title='Date',
    yaxis_title='Counts',
    plot_bgcolor='#FAF9F6',
    xaxis_rangeslider_visible=True,
    xaxis=dict(showgrid=True),
    yaxis=dict(showgrid=True),
    height=700,
    width= 1000,
)

pyo.iplot(fig)
```

From the plot, we can see that the highest volume of tweets was generated in the later part of the pre_season (2741 tweets on July 19) when Manchester City lost the FA Cup semi final to Arsenal . The tweets fluctuated but dropped to their lowest in September (25 tweets on September 19). Positive tweets were more frequent generally over the time period  and always outnumbered negative tweets. We can argue that overall, people spoke positively about Manchester City on any given day than they did negatively. It is important to note that negative tweets may also capture part of Manchester City's fan base given the emotions carried after losing a game or exiting a competition. Some people's tweets might feature frequently over this time period which means that this is by no means an indicator of the number of fans Manchester City has, coupled with the fact that even rivals or neutrals also tweet about the club.

### Wordcloud

I also generated a wordcloud so that I could see what words featured frequently in the tweets and gain insights on some of the conversations.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
from nltk.util import ngrams
import emoji
import unicodedata
import os, wordcloud
```

```python
#create a string from all tweets
allTweets = ""
for index, row in city_tweets.iterrows():
    allTweets += str(row['text'])

# function to remove emojis
def remove_emojis(text):
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

# apply the function to the 'text' column
allTweets = remove_emojis(allTweets)

random.seed(42)

stop_words = set(stopwords.words('english'))
custom_stopwords = ['https', 'mancity', 'man', 'city', 'co', 'city_xtra', 'manchester', 'mcfc', 'rt', 'xi', 'manchestercity', 'amp']
stop_words.update(custom_stopwords)
tokens = re.findall('\w+', allTweets.lower())
cleanTokens = [token for token in tokens if not token in stop_words]
cleanTweets = ' '.join(cleanTokens)

colors = ['#6CABDD', '#1C2C5B', '#FFC659', '#D4A12A', '#582355', '#1c204e']

wc = wordcloud.WordCloud(width = 800, height = 800, 
                background_color ='white', max_font_size=80, max_words=100, color_func=lambda *args, **kwargs: colors[random.randint(0, len(colors)-1)]).generate(cleanTweets)
plt.figure(figsize=(10, 10))
plt.axis("off")
plt.imshow(wc, interpolation="bilinear")
```

From the word cloud we can see that some of the more popular names and words that featured are Ferran Torres, (Kevin) de Bruyne, Pep Guardiola, David Silva and Lionel Messi. Some club names were also part of the tweets including Real Madrid, Barcelona, Chelsea and Valencia. Some football terms also showed up such as FA Cup, Chamions League,  players transfer and away kit.
